```python
import requests
import sqlalchemy as alch
import os
import dotenv
from getpass import getpass
import pandas as pd
import pymysql
dotenv.load_dotenv()
```

```python
# let's build a connection to the table

pass_w = getpass("Please enter your password: ")
```

```python
dbName = "TED"
connectionData=f"mysql+pymysql://root:{pass_w}@localhost/{dbName}"
engine = alch.create_engine(connectionData)
engine
```

```python
# Add data to an SQL table - item by item

# build the connection
connection = pymysql.connect(host='localhost',
                             user='root',
                             password=pass_w,
                             db='TED')

cursor=connection.cursor()
```

```python
#getting TED talks from 2006 - 2020
TED=pd.read_csv('data/TED.csv')
TED
```

```python
TED.to_sql('Talks', con=engine, if_exists='append')
```

```python
query = "SELECT * FROM Talks"
df = pd.read_sql_query(query, engine)
df
```
